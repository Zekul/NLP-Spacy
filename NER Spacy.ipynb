```python
# We will be working with Spacy to explore named entity recognition
# This is a Machine Learning (NLP) process under Data Mining
```

```python
#importing my modules
import spacy
nlp = spacy.load('en_core_web_sm')  
```

```python
#locating my text file 
with open (r"D:\! data science\data mining\exercise\sampleText2.txt", encoding='utf-8') as f:
    text = f.read()
doc = nlp(text)
#print(text)
```

```python
#tokenization of text 
for ent in doc.ents:
    print(ent.text, ent.start_char, ent.end_char, ent.label_)

print("######################################################")

for token in doc:
    tokens = token.text + ',' + token.lemma_+ ',' + token.pos_ + ',' + token.tag_ + ',' + token.dep_ + ',' +\
             token.shape_ + ',' +  str(token.is_alpha) + ',' + str(token.is_stop) + '\n'
print(tokens)
```

```python
# Wow that is alot of tokenized words, Tokenization is breaking the article
# into useful "tokens" which will help us to extract the key information later
# Below is a visual demonstration of the text being tokenized
```

```python
#get visualization of spacy tokenizations
from spacy import displacy
displacy.render(doc, style="ent")
```

```python
#counting all the labels
from collections import Counter
labels = [x.label_ for x in doc.ents]
Counter(labels)
```

```python
# For the purpose of this demonstration I want to extract the date
```

```python
date_list = []
for ent in doc.ents:
    if ent.label_ == 'DATE':
        date_list.append(ent.text)
print(date_list)
```

```python
# The data will look neater in a dataframe that can summarize the mention of certain keywords
```

```python
import pandas as pd
from collections import Counter
```

```python
date_counts = Counter(date_list).most_common()
df_date = pd.DataFrame(date_counts, columns = ['dates', 'count'])

print(df_date)
```
